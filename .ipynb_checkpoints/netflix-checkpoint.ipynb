```python
import pandas as pd
```

```python
# Read the dataset into a pandas DataFrame named netflix_titles.
read_data = pd.read_csv('data/netflix_titles.csv')
netflix_titles_df = pd.DataFrame(read_data)
netflix_titles_df
```

```python
# Display the first 5 rows of the DataFrame.
netflix_titles_df.head()
```

```python
# Show basic information about the dataset (e.g., number of rows/columns, data types).
print(netflix_titles_df.info())
```

```python
# Count the number of titles per country.
number_titles = netflix_titles_df['title'].groupby(netflix_titles_df['country']).count()
print("<----- Number of titles per country ----->")
print(number_titles)
# Find the top 5 countries in terms of the number of titles.
top_five_countries = number_titles.sort_values(ascending=False).head(5)
print("<----- Top Five Countries ----->")
print(top_five_countries)
# Identify the year with the highest number of releases.
year_highest_release = netflix_titles_df['release_year'].groupby(netflix_titles_df['release_year']).count().sort_values(ascending=False).head(1)
print("<----- The year with the highest number of releases. ----->")
print(year_highest_release)
```

```python
# Create a new DataFrame named movies that contains only the entries where the type is 'Movie', including all relevant columns from the original dataset.
movie_df = netflix_titles_df[netflix_titles_df["type"] == "Movie"]
movie_df
```

```python
# List unique directors present in the dataset.
unique_directors = netflix_titles_df['director'].unique()
print("<----- Unique directors ----->")

print(unique_directors)
```

```python
# Replace missing director values with "unknown".
replace_missing_director = netflix_titles_df[netflix_titles_df['director'].fillna("unknown")]
print("<----- Replace missing director values with \"unknown\" ----->")
replace_missing_director
```
