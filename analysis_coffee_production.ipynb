```python
import pandas as pd
```

```python
path_file = 'data/coffee_production.csv'
coffee_production_df = pd.read_csv(path_file)
coffee_production_df.head()
```

```python
# Group the data and summarize the production of each coffee type
group_data = coffee_production_df.groupby('coffee_type').sum()
group_data
```

```python
# Transpose the data
transpose_data = group_data.T
transpose_data
```

```python
# Correlation data
corr_matrix = transpose_data.corr(method='pearson').abs()
corr_matrix
```

```python
# The strongest correlation.  Which two coffee types have the strongest correlation in production volumes over the years? 
strongest_corr = corr_matrix[corr_matrix != 1].unstack().sort_values(ascending=False).drop_duplicates()
strongest_corr.head(1)
```

```python
# The weakest correlation.  Which two coffee types have the weakest correlation in production volumes over the years?
weakest_corr = corr_matrix[corr_matrix != 1].unstack().sort_values(ascending=True).drop_duplicates()
weakest_corr.head(1)
```

```python
# Visualization
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 8))
# show the correlation matrix
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```
